Describe the parameters and quantum state space.

\begin{align*}
    n &= \text{number of qubits} \\
    N &= \text{number of samples} \\
    S &= \text{shape of output array} = (m_1, m_2) \in \mathbb{N}^2
\end{align*} \\



 Procedure for creating quantum circuits,
\begin{align*}
    U_{QC}(n,\epsilon,\nu) : \mathcal{H}_n \rightarrow \mathcal{H}_n
\end{align*} \\




Distribution of probabilities over measurement results,
\begin{align*}
    P(x|\psi) = |\langle x | U_{QC}(n,\epsilon,\nu) |\psi\rangle|^2 \quad x \in \{0,1\}^n
\end{align*} \\




 Process of parallel circuit execution,
\begin{align*}
    P_i(x|\psi) &= |\langle x | U^{(i)}_{QC}(n,\epsilon,\nu) |\psi\rangle|^2 \\
    i &\in \{1,2,3\}
\end{align*} \\


The combined probability distribution,

\begin{align*}
    P_c(x) = \frac{1}{N}\sum_{i=1}^3 \sum_{j=1}^N \delta(x,x_j^{(i)})
\end{align*} \\


 The practice of creating noise patterns,
\begin{align*}
    \eta(m_1,m_2) = \frac{\xi(m_1m_2) - \frac{2^n-1}{2}}{\frac{2^n-1}{2}}
\end{align*} \\



\begin{align*}
    \eta'(m_1,m_2) = \frac{\eta(m_1,m_2) - \mathbb{E}[\eta]}{\sqrt{\text{Var}[\eta]}}
\end{align*} \\

```python
# @title

# Import Aer from qiskit_aer for backend
from qiskit_aer import Aer
# from qiskit import execute # Keep execute for now if needed for basic Aer simulation
from qiskit_aer import AerSimulator # Import AerSimulator

import numpy as np
from art.estimators.classification import KerasClassifier
# Correct import path for attacks
from art.attacks.evasion import FastGradientMethod, BoundaryAttack, BasicIterativeMethod, ProjectedGradientDescent, NewtonFool # Corrected import path

from art.attacks.inference.membership_inference import MembershipInferenceBlackBox
from art.attacks.evasion import HopSkipJump
from art.attacks.evasion import ProjectedGradientDescent

import torch
from art.estimators.classification import KerasClassifier
from art.attacks.evasion import FastGradientMethod, BoundaryAttack # Ensure these are imported correctly
from tensorflow.keras.datasets import mnist # Import mnist to load data
from tensorflow.keras.models import Sequential, Model # Import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation, Add, Input # Import necessary layers
from tensorflow.keras.optimizers import Nadam # Import Nadam optimizer
from tensorflow.keras import regularizers # Import regularizers
from tensorflow.keras import backend as K # Import Keras backend


# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Create classical neural network model


def create_model():
    def residual_block(x, filters, kernel_size=(3, 3)):
        # Store the input to the block for the shortcut connection
        shortcut = x

        y = Conv2D(filters, kernel_size, padding='same')(x)
        y = BatchNormalization()(y)
        y = Activation('relu')(y)
        y = Conv2D(filters, kernel_size, padding='same')(y)
        y = BatchNormalization()(y)

        # Match the number of filters and spatial dimensions for the shortcut connection
        if K.int_shape(shortcut)[-1] != filters or K.int_shape(shortcut)[1:3] != K.int_shape(y)[1:3]:
             # Use a 1x1 convolution with appropriate stride to match shapes
             # Determine stride based on spatial dimension change
             stride = 1
             if K.int_shape(shortcut)[1] != K.int_shape(y)[1]:
                  stride = 2 # AssumingMaxPooling2D reduced spatial dimensions by a factor of 2

             shortcut = Conv2D(filters, (1, 1), strides=(stride, stride), padding='same')(shortcut)
             shortcut = BatchNormalization()(shortcut) # Optional: Add batch norm to the shortcut


        y = Add()([y, shortcut]) # Add the output to the modified shortcut
        return Activation('relu')(y)

    # Use functional API for residual blocks
    x_input = Input(shape=(28, 28, 1))
    x = Conv2D(32, (3, 3), padding='same')(x_input)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Residual Blocks
    x = residual_block(x, 32)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    # Feature Extraction
    # The input x to this block will have 32 filters and reduced spatial dimensions due to MaxPooling2D
    # The block will output layers with 64 filters
    x = residual_block(x, 64)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    # Dense Layers with Attention
    x = Flatten()(x)
    x = Dense(128, kernel_regularizer=regularizers.L1(0.01))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)

    # Output
    outputs = Dense(10, activation='softmax')(x)

    model = Model(inputs=x_input, outputs=outputs)

    # Advanced Optimization
    optimizer = Nadam(learning_rate=0.002) # Removed schedule_decay
    model.compile(loss='sparse_categorical_crossentropy',
                  optimizer=optimizer,
                  metrics=['accuracy']) # Removed 'top_k_categorical_accuracy'
    return model

# Create quantum circuit for quantum-inspired perturbations

from qiskit import QuantumCircuit, transpile
import numpy as np


import numpy as np
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.circuit.library import EfficientSU2
from qiskit.quantum_info import Statevector
from qiskit.transpiler.passes import Optimize1qGatesDecomposition
from typing import Optional, List, Dict, Tuple


class QuantumDiffusionModel:
    """
    Class implementing quantum circuit synthesis using diffusion models.
    This implementation follows recent research on quantum circuit synthesis with diffusion models.

    Attributes:
        num_qubits (int): Number of qubits in the circuit
        num_layers (int): Number of layers in the circuit
        available_gates (list): List of supported quantum gates
        device (str): Device to run the model on ('cpu' or 'cuda')
    """
    def __init__(self, num_qubits: int = 8, num_layers: int = 4, device: str = 'cpu'):
        self.num_qubits = num_qubits
        self.num_layers = num_layers
        self.available_gates = ['h', 'cx', 'ccx', 'swap', 'rx', 'ry', 'rz', 'cp']
        self.device = device

        # Initialize parameters for diffusion process
        self.diffusion_steps = 50
        self.guidance_scale = 7.5

        # Set random seed for reproducibility
        torch.manual_seed(42)

    def _encode_prompt(self, prompt: str) -> torch.Tensor:
        """
        Encodes the text prompt into a tensor format suitable for the diffusion model.

        Args:
            prompt (str): Text description of desired quantum circuit

        Returns:
            torch.Tensor: Encoded prompt tensor
        """
        tokenizer = AutoTokenizer.from_pretrained('openai/clip-vit-base-patch32')
        inputs = tokenizer(prompt, return_tensors='pt', max_length=512, truncation=True)
        return inputs['input_ids'].to(self.device)

    def _generate_circuit_tensor(self, prompt_tensor: torch.Tensor) -> torch.Tensor:
        """
        Generates a circuit tensor representation using diffusion model.

        Args:
            prompt_tensor (torch.Tensor): Encoded prompt tensor

        Returns:
            torch.Tensor: Circuit representation tensor
        """
        # Initialize diffusion pipeline
        pipe = StableDiffusionPipeline.from_pretrained(
            'stabilityai/stable-diffusion-3.5-large',  #CompVis/stable-diffusion-v1-4; stabilityai/stable-diffusion-3.5-large; CompVis/stable-diffusion-v1-inference
            device_map="auto"
        )

        # Generate circuit tensor
        with torch.no_grad():
            outputs = pipe(prompt_tensor,
                         num_inference_steps=self.diffusion_steps,
                         guidance_scale=self.guidance_scale)

        return outputs.images[0]

    def generate_circuit(self, prompt: str) -> QuantumCircuit:
        """
        Generates a quantum circuit based on a text prompt using diffusion models.

        Args:
            prompt (str): Text description of desired quantum circuit

        Returns:
            QuantumCircuit: Generated quantum circuit
        """
        # Encode prompt
        prompt_tensor = self._encode_prompt(prompt)

        # Generate circuit tensor
        circuit_tensor = self._generate_circuit_tensor(prompt_tensor)

        # Convert tensor to quantum circuit
        qr = QuantumRegister(self.num_qubits, 'q')
        cr = ClassicalRegister(self.num_qubits, 'c')
        qc = QuantumCircuit(qr, cr)

        # Decode tensor into gates
        gate_probabilities = torch.nn.functional.softmax(circuit_tensor, dim=-1)

        for layer in range(self.num_layers):
            for qubit_idx in range(self.num_qubits):
                gate_idx = torch.argmax(gate_probabilities[layer, qubit_idx])
                gate_name = self.available_gates[gate_idx]

                if gate_name == 'rx':
                    qc.rx(np.pi/2, qr[qubit_idx])
                elif gate_name == 'ry':
                    qc.ry(np.pi/2, qr[qubit_idx])
                elif gate_name == 'rz':
                    qc.rz(np.pi/2, qr[qubit_idx])
                elif gate_name == 'h':
                    qc.h(qr[qubit_idx])

            # Add entangling gates
            for i in range(self.num_qubits-1):
                if torch.random.random() > 0.5:
                    qc.cx(qr[i], qr[i+1])

        return qc

def _integrate_diffusion_model(qc: QuantumCircuit, qr: QuantumRegister, diffusion_model: QuantumDiffusionModel) -> None:
    """
    Integrates a quantum diffusion model into the quantum circuit.

    Args:
        qc (QuantumCircuit): The quantum circuit to integrate into
        qr (QuantumRegister): The quantum register containing the qubits
        diffusion_model (QuantumDiffusionModel): The diffusion model instance

    Returns:
        None: Modifies the circuit in-place
    """
    # Get the number of qubits needed for the diffusion model
    num_qubits_needed = diffusion_model.num_qubits

    # Ensure we have enough qubits available
    if len(qr) < num_qubits_needed:
        raise ValueError(f"Circuit requires at least {num_qubits_needed} qubits, "
                         f"but only {len(qr)} qubits are available")

    # Select the qubits to use for the diffusion model
    selected_qubits = qr[:num_qubits_needed]

    # Generate a template circuit from the diffusion model
    template_circuit = diffusion_model.generate_circuit("template")

    # Convert the template circuit to an instruction for easier manipulation
    template_instruction = template_circuit.to_instruction()

    # Add the diffusion model circuit as a sub-circuit
    qc.append(template_instruction, selected_qubits)

    # Add error correction checks around the diffusion model integration
    for i in range(num_qubits_needed):
        # Add error check before the diffusion model
        _add_error_check(qc, qr, 'anc', i)

        # Apply the diffusion model
        qc.barrier(selected_qubits)

        # Add error check after the diffusion model
        _add_error_check(qc, qr, 'anc', i)


def create_quantum_circuit(num_qubits: int = 8,
                          num_classical_bits: Optional[int] = None,
                          error_correction: bool = True,
                          noise_simulation: bool = False,
                          optimization_level: int = 3,
                          diffusion_model: Optional['QuantumDiffusionModel'] = None) -> QuantumCircuit:
    """
    Creates a realistic quantum circuit with error correction, noise resilience, and quantum diffusion model integration.

    Args:
        num_qubits (int): Number of qubits in the circuit
        num_classical_bits (Optional[int]): Number of classical bits for measurement
        error_correction (bool): Whether to include error correction
        noise_simulation (bool): Whether to simulate realistic noise
        optimization_level (int): Qiskit optimization level (0-3)
        diffusion_model (Optional[QuantumDiffusionModel]): Quantum diffusion model instance

    Returns:
        QuantumCircuit: Optimized quantum circuit with realistic considerations
    """
    # Set number of classical bits if not specified
    num_classical_bits = num_qubits if num_classical_bits is None else num_classical_bits

    # Initialize registers
    qr = QuantumRegister(num_qubits, 'q')
    cr = ClassicalRegister(num_classical_bits, 'c')
    ancilla = QuantumRegister(2, 'anc')

    # Create base circuit
    qc = QuantumCircuit(qr, ancilla, cr)

    # Add dynamical decoupling pulses for noise reduction
    if noise_simulation:
        _add_dynamical_decoupling(qc, qr)

    # Implement surface code pattern for error correction
    if error_correction:
        _apply_surface_code_pattern(qc, qr)

    # Apply Hadamard gates for superposition with error correction
    for i in range(num_qubits):
        qc.h(qr[i])

        # Add error correction checks after each gate
        if error_correction:
            _add_error_check(qc, qr, ancilla, i)

    # Implement controlled rotations with error correction
    for i in range(num_qubits-1):
        qc.crx(np.pi/4, qr[i], qr[num_qubits-1])

        # Add error correction checks after CRX gates
        if error_correction:
            _add_error_check(qc, qr, ancilla, num_qubits-1)

    # Integrate quantum diffusion model if provided
    if diffusion_model:
        _integrate_diffusion_model(qc, qr, diffusion_model)

    # Add measurement error mitigation
    _add_measurement_error_mitigation(qc, qr, cr)

    _apply_surface_code_pattern(qc, qr)

    # Optimize the circuit
    qc = transpile(qc, optimization_level=optimization_level)

    return qc


def _add_dynamical_decoupling(qc: QuantumCircuit, qr: QuantumRegister) -> None:
    """Adds dynamical decoupling pulses to reduce noise"""
    pulse_sequence = ['X', 'Y', '[-X]', '[-Y]']
    for gate in pulse_sequence:
        for qubit in qr:
            if gate.startswith('[-'):
                qc.x(qubit).inverse()
            elif gate == 'X':
                qc.x(qubit)
            elif gate == 'Y':
                qc.y(qubit)


def _apply_surface_code_pattern(qc: QuantumCircuit, qr: QuantumRegister) -> None:
    """Implements surface code pattern for error correction"""
    for i in range(0, len(qr)-1, 2):
        qc.cz(qr[i], qr[i+1])
        if i < len(qr)-2:
            qc.cx(qr[i], qr[i+2])
            qc.cx(qr[i+1], qr[i+2])
            qc.cx(qr[i], qr[i+2])

def _add_error_check(qc: QuantumCircuit, qr: QuantumRegister, ancilla: QuantumRegister, qubit_index: int) -> None:
    """Adds error checking after quantum operations"""
    # Measure syndrome bits for error detection
    qc.ccx(qr[qubit_index], qr[(qubit_index+1)%len(qr)], ancilla[0])
    qc.cx(qr[qubit_index], ancilla[1])
    qc.ccx(qr[(qubit_index+1)%len(qr)], qr[(qubit_index+2)%len(qr)], ancilla[0])
    qc.cx(qr[(qubit_index+1)%len(qr)], ancilla[1])
    qc.ccx(ancilla[0], ancilla[1], qr[(qubit_index+2)%len(qr)])

def _add_measurement_error_mitigation(qc: QuantumCircuit, qr: QuantumRegister, cr: ClassicalRegister) -> None:
    """Adds measurement error mitigation techniques"""
    for i in range(len(qr)):
        qc.measure(qr[i], cr[i])



def generate_quantum_perturbations(circuit: QuantumCircuit,
                                 num_samples: int = 100,
                                 noise_model: Optional[str] = None) -> Dict[str, float]:
    """
    Generates quantum perturbations using AerSimulator.

    Args:
        circuit (QuantumCircuit): Quantum circuit to simulate
        num_samples (int): Number of shots for simulation
        noise_model (Optional[str]): Noise model to apply

    Returns:
        Dict[str, float]: Probabilities of measurement outcomes
    """
    backend = AerSimulator(noise_model=noise_model)
    job = backend.run(circuit, shots=num_samples)

    counts = job.result().get_counts(circuit)
    probabilities = {k: v/num_samples for k, v in counts.items()}
    return probabilities
```

```python
# @title


#Good code
class QuantumNoiseGenerator:
    def __init__(self, num_qubits=8, error_correction=True, noise_simulation=True):
        self.num_qubits = num_qubits
        self.error_correction = error_correction
        self.noise_simulation = noise_simulation
        self.backend = AerSimulator()
        self.circuit = create_quantum_circuit(
            num_qubits=self.num_qubits,
            error_correction=self.error_correction,
            noise_simulation=self.noise_simulation
        )

    def generate_quantum_noise(self, shape, num_samples=1000):
        """Generate quantum noise patterns using actual quantum circuits."""
        backend = AerSimulator()

        # Create multiple circuits for different noise patterns
        circuits = [create_quantum_circuit(
            num_qubits=self.num_qubits,
            error_correction=self.error_correction,
            noise_simulation=self.noise_simulation
        ) for _ in range(3)]  # Multiple circuits for diverse patterns

        # Run all circuits
        jobs = [backend.run(circuit, shots=num_samples) for circuit in circuits]
        results = [job.result().get_counts(circuit)
                  for job, circuit in zip(jobs, circuits)]

        # Combine results for more complex patterns
        combined_probs = {}
        for result in results:
            for key, value in result.items():
                combined_probs[key] = combined_probs.get(key, 0) + value

        # Normalize probabilities
        total_shots = sum(combined_probs.values())
        combined_probs = {k: v/total_shots for k, v in combined_probs.items()}

        # Generate noise pattern based on quantum measurements
        flat_shape = shape[0] * shape[1]
        noise_pattern = np.random.choice(
            list(combined_probs.keys()),
            size=flat_shape,
            p=list(combined_probs.values())
        ).reshape(shape)

        # Scale noise appropriately
        max_bits = len(list(combined_probs.keys())[0])
        scaled_noise = (noise_pattern.astype(float) / (2**max_bits - 1) - 0.5)

        return scaled_noise
```

```python
# @title

def quantum_inspired_fgsm(model, images, eps=0.3):
    classifier = KerasClassifier(model=model)
    fgsm = FastGradientMethod(classifier, eps=eps)

    # Initialize quantum noise generator
    qng = QuantumNoiseGenerator(num_qubits=8)

    adversarial_images = []
    for img in images:
        # Generate classical FGSM attack
        adv_img = fgsm.generate(x=img[np.newaxis])[0]

        # Generate quantum-inspired perturbation
        quantum_noise = qng.generate_quantum_noise(img.shape)

        # Combine attacks with weighted contribution
        final_adv = adv_img + 0.5 * eps * quantum_noise

        # Ensure pixel values stay within bounds
        final_adv = np.clip(final_adv, 0, 1)
        adversarial_images.append(final_adv)

    return np.array(adversarial_images)
```

Quantum Boundary Attack,  
\begin{equation*}
\begin{aligned}
x^{\text{adv}} &= \text{clip}_{[0,1]}\left(x^{\text{base}} + \omega\mathcal{N}_q\right) \\
x^{\text{base}} &= \arg\min_{x'} \|\nabla L(f_{\theta}(x'), y)\|_2 \\
\mathcal{N}_q &= \mathcal{Q}_{\omega}(x \in [0,1]^n)
\end{aligned}
\end{equation*}



Quantum Basic Iterative Method,  
\begin{equation*}
\begin{aligned}
x^{k+1} &= \text{clip}_{[0,1]}(x^k + \alpha\text{sign}(g^k)) \\
g^k &= \nabla L(f_{\theta}(x^k), y) + \omega\mathcal{N}_q \\
\mathcal{N}_q &= \mathcal{Q}_{\omega}(x \in [0,1]^n)
\end{aligned}
\end{equation*} \\





Quantum NewtonFool,  
\begin{equation*}
\begin{aligned}
H(x) &= \frac{\partial^2 L(f_{\theta}(x), y)}{\partial x^2} \\
g(x) &= \nabla L(f_{\theta}(x), y) + \omega\mathcal{N}_q \\
x^{k+1} &= x^k - \eta(H^{-1}g)^k
\end{aligned}
\end{equation*}


 Quantum HopSkipJump,  
\begin{equation*}
\begin{aligned}
x^{k+1} &= x^k + \beta_k\text{sign}(g^k) \\
g^k &= \nabla L(f_{\theta}(x^k), y) + \omega\mathcal{N}_q \\
\beta_k &= \min(\beta_{\max}, \beta_0 + k\Delta\beta)
\end{aligned}
\end{equation*} \\



 Quantum Noise Injection,  
\begin{equation*}
\begin{aligned}
\mathcal{N}_q &= \mathcal{Q}_{\omega}(x) \\
&= \text{reshape}(\mathcal{Q}'(q=n), (1,1) + \text{shape}(x))
\end{aligned}
\end{equation*}

```python
# @title

class QuantumBoundaryAttack(BoundaryAttack):
    def __init__(self, estimator, targeted=False, max_iter=100,
                 quantum_weight=0.5):
        super().__init__(
            estimator=estimator,
            targeted=targeted,
            max_iter=max_iter

        )
        self.quantum_weight = quantum_weight
        self.qng = QuantumNoiseGenerator(num_qubits=8)

    def generate(self, x, y=None):
        """Generate Boundary attacks with quantum-inspired perturbations."""
        import tensorflow as tf

        # Get initial point
        x_adv = super().generate(x, y)

        # Add quantum noise to the boundary points
        quantum_noise = self.qng.generate_quantum_noise(x.shape[1:])
        # Reshape quantum noise for broadcasting with boundary points
        quantum_noise_reshaped = np.reshape(quantum_noise, (1,) + quantum_noise.shape)
        x_adv = tf.clip_by_value(
            x_adv + self.quantum_weight * quantum_noise_reshaped,
            clip_value_min=0,
            clip_value_max=1
        )

        return x_adv



class QuantumBasicIterativeMethod(BasicIterativeMethod):
    def __init__(self, estimator, eps=0.3, eps_step=0.1,
                 max_iter=40, targeted=False,
                 quantum_weight=0.5):
        super().__init__(
            estimator=estimator,
            eps=eps,
            eps_step=eps_step,
            max_iter=max_iter,
            targeted=targeted

        )
        self.quantum_weight = quantum_weight
        self.qng = QuantumNoiseGenerator(num_qubits=8)

    def generate(self, x, y=None):
        """Generate BIM attacks with quantum-inspired perturbations."""
        import tensorflow as tf

        # Initialize best result
        best_adv = tf.identity(x)

        # Try multiple random initializations
        for i in range(self.num_random_init):
            # Generate random initialization within epsilon ball
            if self.targeted:
                random_perturbation = tf.random.uniform(tf.shape(x),
                                                      minval=-self.eps,
                                                      maxval=self.eps)
            else:
                random_perturbation = tf.random.uniform(tf.shape(x),
                                                      minval=-self.eps,
                                                      maxval=self.eps)
            current = tf.clip_by_value(x + random_perturbation, 0, 1)

            # Perform iterative attack with quantum noise
            for _ in range(self.max_iter):
                # Get gradient with quantum noise injection
                grad = self.estimator.class_gradient(current, y=y)
                quantum_noise = self.qng.generate_quantum_noise(x.shape[1:])
                # Reshape quantum_noise to match gradient shape for broadcasting
                quantum_noise_reshaped = np.reshape(quantum_noise, (1, 1) + quantum_noise.shape)
                noisy_grad = grad + self.quantum_weight * tf.convert_to_tensor(quantum_noise_reshaped)

                # Update adversarial example
                if not self.targeted:
                    current = tf.clip_by_value(
                        current + self.eps_step * tf.sign(noisy_grad),
                        clip_value_min=0,
                        clip_value_max=1
                    )
                else:
                    current = tf.clip_by_value(
                        current - self.eps_step * tf.sign(noisy_grad),
                        clip_value_min=0,
                        clip_value_max=1
                    )

                # Project back onto feasible region
                if not self.targeted:
                    current = tf.clip_by_value(
                        current,
                        clip_value_min=x - self.eps,
                        clip_value_max=x + self.eps
                    )
                else:
                    current = tf.clip_by_value(
                        current,
                        clip_value_min=x - self.eps,
                        clip_value_max=x + self.eps
                    )

            # Update best result
            if self._compare(x, best_adv, current):
                best_adv = current

        return best_adv


class QuantumNewtonFool(NewtonFool):
    def __init__(self, classifier, max_iter=100, eta=0.03,
                 clip_min=None, clip_max=None, batch_size=128,
                 verbose=False, quantum_weight=0.5):
        super().__init__(
            classifier=classifier,
            max_iter=max_iter,
            eta=eta,
            batch_size=batch_size,
            verbose=verbose
        )
        self.quantum_weight = quantum_weight
        self.qng = QuantumNoiseGenerator(num_qubits=8)

    def _compute_gradient_hessian(self, x, y):
        """Compute gradient and Hessian with quantum noise injection."""
        import tensorflow as tf

        # Compute gradient
        grad = self.estimator.class_gradient(x, y=y)

        # Add quantum noise to gradient
        quantum_noise = self.qng.generate_quantum_noise(x.shape[1:])
        # Reshape quantum_noise to match grad shape for broadcasting
        quantum_noise_reshaped = np.reshape(quantum_noise, (1, 1) + quantum_noise.shape)
        noisy_grad = grad + self.quantum_weight * tf.convert_to_tensor(quantum_noise_reshaped)

        # Compute Hessian-vector product
        def hvp(v):
            # Hessian-vector product also needs target labels
            return self.estimator.hessian_vector_product(x, v, y=y)

        return noisy_grad, hvp


class QuantumHopSkipJump(HopSkipJump):
    def __init__(self, classifier, targeted=False, max_iter=64,
                 max_eval=10000, init_eval=100, init_size=32,
                 batch_size=128, quantum_weight=0.5):
        super().__init__(
            classifier=classifier,
            targeted=targeted,
            max_iter=max_iter,
            max_eval=max_eval,
            init_eval=init_eval,
            init_size=init_size,
            batch_size=batch_size
        )
        self.quantum_weight = quantum_weight
        self.qng = QuantumNoiseGenerator(num_qubits=8)

    def _compute_gradient(self, x, y):
        """Compute gradient with quantum noise injection."""
        grad = super()._compute_gradient(x, y)
        quantum_noise = self.qng.generate_quantum_noise(x.shape[1:])
        # Reshape quantum_noise to match grad shape for broadcasting
        quantum_noise_reshaped = np.reshape(quantum_noise, (1, 1) + quantum_noise.shape)
        return grad + self.quantum_weight * tf.convert_to_tensor(quantum_noise_reshaped)

def evaluate_all_attacks(model, x_test, y_test):
    """Evaluate all three attack methods with quantum integration."""
    classifier = KerasClassifier(model=model)

    # Initialize attacks
    qba = QuantumBoundaryAttack(estimator=classifier, max_iter=25)
    qhsj = QuantumHopSkipJump(classifier=classifier, max_iter=25)
    qi_fgsm = FastGradientMethod(classifier, eps=0.5)
    qng = QuantumNoiseGenerator(num_qubits=8)

    print("\nAttack Evaluation Results:")

    # Clean accuracy
    loss, accuracy = model.evaluate(x_test[:100], y_test[:100], verbose=0)
    print(f"Clean accuracy: {accuracy:.2f}")

    # Quantum-Inspired FGSM
    fgsm_attacks = []
    for img in x_test[:100]:
        adv_img = qi_fgsm.generate(x=img[np.newaxis])[0]
        quantum_noise = qng.generate_quantum_noise(img.shape)
        final_attack = adv_img + 0.5 * 0.3 * quantum_noise
        final_attack = np.clip(final_attack, 0, 1)
        fgsm_attacks.append(final_attack)


    loss, accuracy = model.evaluate(np.array(fgsm_attacks), y_test[:100], verbose=0)
    print(f"Quantum-Inspired FGSM accuracy: {accuracy:.2f}")

    # Quantum-Inspired HopSkipJump
    hsja_attacks = qhsj.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(hsja_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired HopSkipJump accuracy: {accuracy:.2f}")

    # Quantum-Inspired Boundary Attack
    ba_attacks = qba.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(ba_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired Boundary accuracy: {accuracy:.2f}")


    # Quantum-Inspired NewtonFool
    qnf = QuantumNewtonFool(classifier=classifier, max_iter=10, eta=0.03)
    nf_attacks = qnf.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(nf_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired NewtonFool accuracy: {accuracy:.2f}")


    # Quantum-Inspired BIM
    qbim = QuantumBasicIterativeMethod(estimator=classifier, eps=0.3, max_iter=40)
    bim_attacks = qbim.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(bim_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired BIM accuracy: {accuracy:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from art.estimators.classification import KerasClassifier
from art.attacks.evasion import FastGradientMethod, BoundaryAttack # Ensure these are imported correctly
from tensorflow.keras.datasets import mnist # Import mnist to load data
from tensorflow.keras.models import Sequential, Model # Import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation, Add, Input # Import necessary layers
from tensorflow.keras.optimizers import Nadam # Import Nadam optimizer
from tensorflow.keras import regularizers # Import regularizers
from tensorflow.keras import backend as K # Import Keras backend


def create_model():
    def residual_block(x, filters, kernel_size=(3, 3)):
        # Store the input to the block for the shortcut connection
        shortcut = x

        y = Conv2D(filters, kernel_size, padding='same')(x)
        y = BatchNormalization()(y)
        y = Activation('relu')(y)
        y = Conv2D(filters, kernel_size, padding='same')(y)
        y = BatchNormalization()(y)

        # Match the number of filters and spatial dimensions for the shortcut connection
        if K.int_shape(shortcut)[-1] != filters or K.int_shape(shortcut)[1:3] != K.int_shape(y)[1:3]:
             # Use a 1x1 convolution with appropriate stride to match shapes
             # Determine stride based on spatial dimension change
             stride = 1
             if K.int_shape(shortcut)[1] != K.int_shape(y)[1]:
                  stride = 2 # AssumingMaxPooling2D reduced spatial dimensions by a factor of 2

             shortcut = Conv2D(filters, (1, 1), strides=(stride, stride), padding='same')(shortcut)
             shortcut = BatchNormalization()(shortcut) # Optional: Add batch norm to the shortcut


        y = Add()([y, shortcut]) # Add the output to the modified shortcut
        return Activation('relu')(y)

    # Use functional API for residual blocks
    x_input = Input(shape=(28, 28, 1))
    x = Conv2D(32, (3, 3), padding='same')(x_input)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Residual Blocks
    x = residual_block(x, 32)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    # Feature Extraction
    # The input x to this block will have 32 filters and reduced spatial dimensions due to MaxPooling2D
    # The block will output layers with 64 filters
    x = residual_block(x, 64)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)

    # Dense Layers with Attention
    x = Flatten()(x)
    x = Dense(128, kernel_regularizer=regularizers.L1(0.01))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)

    # Output
    outputs = Dense(10, activation='softmax')(x)

    model = Model(inputs=x_input, outputs=outputs)

    # Advanced Optimization
    optimizer = Nadam(learning_rate=0.002) # Removed schedule_decay
    model.compile(loss='sparse_categorical_crossentropy',
                  optimizer=optimizer,
                  metrics=['accuracy']) # Removed 'top_k_categorical_accuracy'
    return model

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.gridspec import GridSpec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def create_visualization(model, x_test, y_test, fgsm_attacks, ba_attacks,
                         hsja_attacks, bim_attacks, nf_attacks):
    """
    Creates a comprehensive visualization of multiple adversarial attack results

    Parameters:
    - model: Keras model
    - x_test: Original test images
    - y_test: Test labels
    - fgsm_attacks: FGSM generated attacks
    - qi_attacks: Quantum-inspired attacks
    - ba_attacks: Boundary attacks
    - hopskipjump_attacks: HopSkipJump attacks
    - pgd_attacks: Projected Gradient Descent attacks
    - bim_attacks: Basic Iterative Method attacks
    - nf_attacks: NewtonFool attacks
    """
    # Set up the style
    plt.style.use('seaborn-v0_8-darkgrid')

    # Create figure with GridSpec for better control
    fig = plt.figure(figsize=(24, 18))
    gs = GridSpec(4, 6, figure=fig)

    def plot_examples(ax, original, attacked, title):
        """Helper function to plot individual attack examples"""
        ax.imshow(original.reshape(28, 28), cmap='gray')

        # Add model prediction if available
        if isinstance(model, Model):
            pred = model.predict(original[np.newaxis,...]).argmax()
            ax.set_title(f'Original\nPred: {pred}')
        else:
            ax.set_title('Original\nModel not available for prediction')

        # Add difference inset
        axins = inset_axes(ax, width="30%", height="30%", loc='upper right')
        if original.shape == attacked.shape:
            diff = np.abs(original - attacked).reshape(28, 28)
            im = axins.imshow(diff, cmap='viridis')
            axins.set_xticks([])
            axins.set_yticks([])
            plt.colorbar(im, ax=axins, location='right', shrink=0.5)
        else:
            axins.text(0.5, 0.5, 'Shape Mismatch',
                      horizontalalignment='center',
                      verticalalignment='center',
                      transform=axins.transAxes)

    # Plot first example comparisons
    if len(x_test) > 0:
        plot_examples(fig.add_subplot(gs[0, 0]), x_test[0], fgsm_attacks[0], 'FGSM Attack')
        plot_examples(fig.add_subplot(gs[0, 2]), x_test[0], ba_attacks[0], 'Boundary Attack')
        plot_examples(fig.add_subplot(gs[0, 3]), x_test[0], hsja_attacks[0], 'HopSkipJump Attack')
        plot_examples(fig.add_subplot(gs[0, 5]), x_test[0], bim_attacks[0], 'BIM Attack')
        plot_examples(fig.add_subplot(gs[0, 5]), x_test[0], nf_attacks[0], 'NewtonFool') # This will overwrite BIM example, need to adjust layout


    # Create accuracy comparison plot
    ax_acc = fig.add_subplot(gs[1, :3]) # Adjusted span
    acc_data = {}

    # Evaluate on clean data if model is available
    if isinstance(model, Model):
        acc_data['Clean'] = model.evaluate(x_test[:100], y_test[:100], verbose=0)[1]
    else:
        acc_data['Clean'] = 0.0

    # Evaluate all attack types
    attack_types = [
        ('FGSM', fgsm_attacks),
        ('Ba', ba_attacks),
        ('BIM', bim_attacks),
        ('NewtonFool', nf_attacks),
        ('HopSkipJump', hsja_attacks)
    ]

    for attack_name, attacks in attack_types:
        if len(attacks) > 0 and attacks.shape == x_test[:100].shape:
            if isinstance(model, Model):
                acc_data[attack_name] = model.evaluate(
                    attacks, y_test[:100], verbose=0)[1]
            else:
                acc_data[attack_name] = 0.0
        else:
            acc_data[attack_name] = 0.0

    sns.barplot(x=list(acc_data.keys()), y=list(acc_data.values()), ax=ax_acc)
    ax_acc.set_ylabel('Accuracy')
    ax_acc.set_ylim(0, 1)
    ax_acc.set_title('Attack Accuracy Comparison') # Added title for clarity
    plt.xticks(rotation=45, ha='right') # Rotate labels for better readability


    def plot_confusion_matrix(ax, predictions, labels, title):
        """Helper function to plot confusion matrices"""
        if predictions is None or predictions.shape[0] != len(labels):
            ax.text(0.5, 0.5, 'No Predictions Available',
                   horizontalalignment='center',
                   verticalalignment='center',
                   transform=ax.transAxes)
            ax.set_title(title)
            return

        conf_mat = np.zeros((10, 10))
        pred_labels = predictions.argmax(axis=-1)
        for i in range(len(labels)):
            if pred_labels[i] < 10:
                conf_mat[labels[i]][pred_labels[i]] += 1

        sns.heatmap(conf_mat, annot=True, fmt='.0f', cmap='YlOrRd',
                   ax=ax, cbar_kws={'label': 'Count'})
        ax.set_xlabel('Predicted Label')
        ax.set_ylabel('True Label')
        ax.set_title(title)

    # Plot confusion matrices for different attacks
    clean_predictions = model.predict(x_test[:100]) if isinstance(model, Model) else None
    attack_predictions = {
        'FGSM': model.predict(fgsm_attacks) if isinstance(model, Model) and len(fgsm_attacks) > 0 else None,
        'Boundary': model.predict(ba_attacks) if isinstance(model, Model) and len(ba_attacks) > 0 else None,
        'HopSkipJump': model.predict(hsja_attacks) if isinstance(model, Model) and len(hsja_attacks) > 0 else None,
        "NewtonFool": model.predict(nf_attacks) if isinstance(model, Model) and len(nf_attacks) > 0 else None,
        "BIM": model.predict(bim_attacks) if isinstance(model, Model) and len(bim_attacks) > 0 else None
    }

    # Plot confusion matrices (adjusting subplot indices)
    plot_confusion_matrix(fig.add_subplot(gs[1, 3:]), clean_predictions, y_test[:100], 'Clean Data Confusion Matrix') # Adjusted subplot
    plot_confusion_matrix(fig.add_subplot(gs[2, :2]), attack_predictions['FGSM'], y_test[:100], 'FGSM Attack Confusion Matrix') # Adjusted subplot
    plot_confusion_matrix(fig.add_subplot(gs[2, 4:]), attack_predictions['Boundary'], y_test[:100], 'Boundary Attack Confusion Matrix') # Adjusted subplot
    plot_confusion_matrix(fig.add_subplot(gs[3, :2]), attack_predictions['HopSkipJump'], y_test[:100], 'HopSkipJump Attack Confusion Matrix') # Adjusted subplot
    plot_confusion_matrix(fig.add_subplot(gs[3, 2:4]), attack_predictions['BIM'], y_test[:100], 'BIM Attack Confusion Matrix') # Adjusted subplot
    plot_confusion_matrix(fig.add_subplot(gs[3, 4:]), attack_predictions['NewtonFool'], y_test[:100], 'NewtonFool Attack Confusion Matrix') # Adjusted subplot

    # Adjust spacing using GridSpec's update method
    gs.update(wspace=0.4, hspace=0.4) # Increased spacing

    return fig

def evaluate_and_visualize():
    """Evaluates attacks and creates visualization"""
    # Load MNIST dataset
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    x_train = x_train.astype('float32') / 255
    x_test = x_test.astype('float32') / 255
    x_train = x_train.reshape(-1, 28, 28, 1)
    x_test = x_test.reshape(-1, 28, 28, 1)

    # Create and train model
    model = create_model()
    model.fit(x_train, y_train, batch_size=128, epochs=2, verbose=1)

    # Generate attacks
    classifier = KerasClassifier(model=model)
    qba = QuantumBoundaryAttack(estimator=classifier, max_iter=30)
    qhsj = QuantumHopSkipJump(classifier=classifier, max_iter=30)
    qi_fgsm = FastGradientMethod(classifier, eps=0.5)
    qng = QuantumNoiseGenerator(num_qubits=8)


    # Quantum-Inspired FGSM
    fgsm_attacks = []
    for img in x_test[:100]:
        adv_img = qi_fgsm.generate(x=img[np.newaxis])[0]
        quantum_noise = qng.generate_quantum_noise(img.shape)
        final_attack = adv_img + 0.5 * 0.3 * quantum_noise
        final_attack = np.clip(final_attack, 0, 1)
        fgsm_attacks.append(final_attack)

    fgsm_attacks = np.array(fgsm_attacks) # Convert list to numpy array


    # Quantum-Inspired HopSkipJump
    hsja_attacks = qhsj.generate(x=x_test[:100], y=y_test[:100],max_iter=30)
    loss, accuracy = model.evaluate(hsja_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired HopSkipJump accuracy: {accuracy:.2f}")

    # Quantum-Inspired Boundary Attack
    ba_attacks = qba.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(ba_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired Boundary accuracy: {accuracy:.2f}")


    # Quantum-Inspired NewtonFool
    qnf = QuantumNewtonFool(classifier=classifier, max_iter=30, eta=0.5)
    nf_attacks = qnf.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(nf_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired NewtonFool accuracy: {accuracy:.2f}")


    # Quantum-Inspired BIM
    qbim = QuantumBasicIterativeMethod(estimator=classifier, eps=0.5, max_iter=40)
    bim_attacks = qbim.generate(x=x_test[:100], y=y_test[:100])
    loss, accuracy = model.evaluate(bim_attacks, y_test[:100], verbose=0)
    print(f"Quantum-Inspired BIM accuracy: {accuracy:.2f}")


    # Create visualization
    fig = create_visualization(
        model, x_test[:100], y_test[:100],
        fgsm_attacks, ba_attacks,
        hsja_attacks, bim_attacks, nf_attacks
    )
    plt.savefig('attack_analysis.png', dpi=300, bbox_inches='tight')
    print("\nDetailed Evaluation Metrics:")
    print("-" * 50)



def calculate_metrics(original, attacked):
     """Helper function to calculate attack metrics"""
     if original.shape != attacked.shape or original.size == 0:
        return {
                'Mean Perturbation': np.nan,
                'Max Perturbation': np.nan,
                'Std Perturbation': np.nan,
                'Median Perturbation': np.nan
            }
     diff = np.abs(original - attacked)
     return {
         'Mean Perturbation': np.mean(diff),
         'Max Perturbation': np.max(diff),
         'Std Perturbation': np.std(diff),
         'Median Perturbation': np.median(diff)
     }

evaluate_and_visualize()
```
